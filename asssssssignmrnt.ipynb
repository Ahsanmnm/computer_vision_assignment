```python
import cv2 as cv
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = 'WhatsApp Image 2022-12-17 at 4.55.08 PM (1).jpeg'
img1 = cv.imread(path)


gray = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

a = int(input())
def intensity_level(a):
    256/(a-1)
    x = round(256/(a-1))
    a =np.round_(gray/x)
    b = a*x
    cv.imshow('img',b)
    cv.waitKey(0)
    cv.destroyAllWindows()
intensity_level(a)  
```

```python
import cv2
img = cv2.imread('Screenshot 2023-01-16 132507.jpg')

kernel = np.ones((3,3),np.float32)/9
dst = cv2.filter2D(img,-1,kernel)

kernell = np.ones((10,10),np.float32)/100
dstt = cv2.filter2D(img,-1,kernell)

kernelll = np.ones((20,20),np.float32)/400
dsttt = cv2.filter2D(img,-1,kernelll)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])


plt.subplot(122),plt.imshow(dst),plt.title('3x3 Averaging')
plt.xticks([]), plt.yticks([])


plt.show()
```

```python
plt.subplot(121),plt.imshow(dstt),plt.title('10x10 Averaging')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(dsttt),plt.title('20x20 Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from PIL import Image

#read the image
im = Image.open("Screenshot 2023-01-16 132507.jpg")

#rotate image
angle = 45
out = im.rotate(angle, expand=True)
plt.imshow(out),plt.title('45 degree')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
image = cv2.imread('Screenshot 2023-01-16 132507.jpg')
height, width = image.shape[:2]
block_size = 3
averaged_image_3_3 = image.copy()
for i in range(0, height-block_size+1, block_size):
    for j in range(0, width-block_size+1, block_size):
        block = image[i:i+block_size, j:j+block_size]
        avg = block.mean()
        averaged_image_3_3[i:i+block_size, j:j+block_size] = avg
fig,arr= plt.subplots(1,2)
fig.subplots_adjust(wspace=0.5) 
arr[0].imshow(image),plt.title('3×3')
arr[1].imshow(averaged_image_3_3)

plt.xticks([]), plt.yticks([])
plt.show()
```
